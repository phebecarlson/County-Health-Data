```python
from sqlalchemy import create_engine

# create sqlalchemy engine
engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"  
                      .format(user="root", pw="cgCbH^1hUx*cFZ5n", 
                      db="countyhealthdb"))

rankingsdf.to_sql('countyhealthrankings', con = engine, if_exists = 'append', chunksize = 1000, index=False)
```
